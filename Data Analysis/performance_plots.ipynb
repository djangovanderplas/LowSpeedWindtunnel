```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 12))
plt.tight_layout(pad=5)
plt.suptitle(f'2D Performance Plots', size=15)

axes[0, 0].plot( df['Alpha'][:32], df['Cl'][:32], marker='.', color='blue')
axes[0, 0].plot( df['Alpha'][31:], df['Cl'][31:], label='Hysteresis', marker='.', color='red')
axes[0, 0].set_xlabel('Alpha [°]')
axes[0, 0].set_ylabel('Cl [-]')
axes[0, 0].set_title('Lift Curve')
axes[0, 0].axhline(y=0, color='k')
axes[0, 0].axvline(x=0, color='k')
axes[0, 0].legend()
axes[0, 0].grid()

plt.show()
```
